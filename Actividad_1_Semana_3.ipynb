Data warehousing
Grandes bancos de datos se tienen que almacenar de forma eficiente para tener resultados optimos tratándose de instituciones públicas o privadas, pues toda esta información se procesa para generar planes de realización para los objetivos específicos de estas instituciones. Lo que pude observar con este tema es que la cantidad de información que se obtiene es verdaderamente grande, por lo tanto hay mucha información que abarca espacio y no tiene utilidad, por lo tanto, los datos se tienen que limpiar.

ETL vs ELT
Como mencioné anteriormente, para un manejo más eficiente de datos que verdaderamente sean útiles, se necesita separarlos de toda la información basura, para eso se tiene que hacer una limpieza. Existen dos opciones:
•	ETL: Que básicamente limpia los datos antes de cargarlos y almacenarlos (forma tradicional).
•	ELT: Que la limpieza de los datos va cuando la información ya está cargada y almacenada (la forma nueva, más eficiente).

Data sources
¿De dónde vienen los datos? Pues básicamente, cualquier interacción con la plataforma digital está generando datos e información, por lo tanto la cantidad es grandísima. Estos datos no se cargan y se almacenan así como vienen, pues para una óptima limpieza de estos se necesita separarlos, los datos numéricos con los de su clase, de igual forma los textos y la multimedia. Como ejemplo de esto está Excel con su función de “Pivot tables”.

Limpieza de datos
En esta parte, por fin vemos como los datos muestran errores o variacionesque se necesitan eliminar, para estos tenemos métodos, pero primero se necesita determinar que para los datos faltantes tenemos 3 tipos: MCAR (Missing Completely at Random), MAR (Missing at Random) y MNAR (Missing Not at Random).
Cada tipo de falta de datos se diferencían por la probabilidad de que falten las respuestas depende del conjunto de respuestas observadas, pero no está relacionada con los valores faltantes específicos que se espera obtener. Para manejar los datos faltantes tenemos las siguientes técnicas:
•	Eliminación por lista
•	Eliminación por parejas
•	Sustitución media
•	Imputación de regresión
•	Imputación Múltiple
